```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Amatrix(a11,a12,a21,a22):
    return np.array([[a11,a12],[a21,a22]])
```

```python
w = interactive(Amatrix,
               a11 = widgets.IntText(
    value=1,
    description=r'$A_{11} = $',
    disabled=False
),
               a12 = widgets.IntText(
    value=7,
    description=r'$A_{12} = $',
    disabled=False
),
               a21 = widgets.IntText(
    value=-2,
    description=r'$A_{21} = $',
    disabled=False
),
               a22 = widgets.IntText(
    value=3,
    description=r'$A_{22} = $',
    disabled=False
))

display(w)
A = w.result
```

```python
#print matrix A
print('A = ['+str(A[0]) +'],'+str(A[1]) +']')

#width of plot
w = 5

#streamplot
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U = A[(0,0)]*X+A[(0,1)]*Y
V = A[(1,0)]*X+A[(1,1)]*Y
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

ax.streamplot(X, Y, U, V)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Stream Plot')
plt.tight_layout()

#vectorfield
Y, X = np.mgrid[-w:w:15j, -w:w:15j]
U = A[(0,0)]*X+A[(0,1)]*Y
V = A[(1,0)]*X+A[(1,1)]*Y

fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)
q = ax.quiver(X, Y, U, V)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Vector Field')
plt.tight_layout()


plt.show()
```
